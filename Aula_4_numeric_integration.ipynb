## Method 1 (Trapezoids): Calculate Integral


```python
def f(x):
  return np.sin(x)

a, b = 0, np.pi
n = 10
h = (b-a)/n

x = np.linspace(a, b, n+1)
y = f(x)

integral = (h/2)*(y[0] +2 *np.sum(y[1:-1]+y[-1]))
print(f"trapezoid:{integral}")
```

```python
# trapezoids
n = 5
x_vals = np.linspace(a, b, n+1)
y_vals = f(x_vals)

x_curve = np.linspace(a, b, 100)
y_curve = f(x_curve)

# plot
plt.plot(x_curve, y_curve, label = 'f(x) = sin(x)')
plt.plot(x_vals, y_vals,'o-', label = 'f(x) = sin(x)',color = 'orange')


plt.show()
```

```python
# fill the trapezoids
for i in range (n):
  xs = [x_vals[i], x_vals[i], x_vals[i+1], x_vals[i+1]]
  ys = [0, y_vals[i], y_vals[i+1], 0]
  plt.fill(xs,ys, color = 'orange' , alpha = 0.5)

plt.title('Rule trapezoid')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

## Method 2: Simpson rule

```python
def simpson(f, a, b, n):

  if n%2 == 1:
    n+=1 # should be pair
    h = (b-a)/n

    x = np.linspace(a, b, n+1)
    y = f(x)

integral = (h/3)*(y[0] + 4*np.sum(y[1:-1:2]) + 2*np.sum(y[2:-2:2]) + y[-1])
print(f"Simpson:{integral}")
```
