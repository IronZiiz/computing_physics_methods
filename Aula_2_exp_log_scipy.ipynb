# Aula 2 Scipy lib




## imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Def Functions

```python
def exponencial(x, a, b, c):
    return a * np.exp(b * x) + c

def logaritimica(x, a, b, c):
  return a * np.log(b * x) + c
```

## Generating data with a bit noise




```python
np.random.seed(42)

x = np.linspace(1, 10, 100)
y_exp = exponencial(x, 2, 0.3, 1) + np.random.normal(0, 0.5, 100)
y_log = logaritimica(x, 2, 0.5, 1) + np.random.normal(0, 0.5, 100)
```

## plot

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y_exp,'o')
plt.title('Exponencial')

plt.subplot(1, 2, 2)
plt.plot(x, y_log,'o')
plt.title('Logarítmica')

plt.show()
```

## Adjusting the curve



```python
params_exp, covariancias_exp = curve_fit(exponencial, x ,y_exp)
params_log, covariancias_log = curve_fit(logaritimica, x ,y_log)

print("Parâmetros da função exponencial:", params_exp)
print("Parâmetros da função logarítmica:", params_log)
```

## plot the better curve

```python
y_exp_fit = exponencial(x, *params_exp)
y_log_fit = logaritimica(x, *params_log)
plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.plot(x, y_exp, 'o', label='Dados')
plt.plot(x, y_exp_fit, '-', label='Ajuste Exponencial')
plt.legend()
plt.title('Exponencial')

plt.subplot(1,2,2)
plt.plot(x, y_log, 'o', label='Dados')
plt.plot(x, y_log_fit, '-', label='Ajuste Exponencial')
plt.legend()
plt.title('Logaritímica')
```

```python

# Calculando o r^2 do modelo exponencial
ss_res_exp = np.sum((y_exp - y_exp_fit)**2)
ss_tot_exp = np.sum((y_exp - np.mean(y_exp))**2)
r2_exp = 1 - (ss_res_exp / ss_tot_exp)

# Calculando o r^2 do modelo log
ss_res_log = np.sum((y_log - y_log_fit)**2)
ss_tot_log = np.sum((y_log - np.mean(y_log))**2)
r2_log = 1 - (ss_res_log / ss_tot_log)

print(f"R-squared EXP : {r2_exp}")
print(f"R-squared LOG: {r2_log}")

```

## Homework
Escolher uma função e fitar e calcular o r²

```python
def sen(x, a=1, b=2, c=3):
    return np.sin(x)*a + b + c
```

```python
np.random.seed(42)

x = np.linspace(1, 10, 100)
y_sin =  sen(x)  + np.random.normal(0, 0.3, 100)

plt.figure(figsize=(12, 5))
plt.plot(x, y_sin,'o')
plt.title('Senoidal')
plt.show()
```

```python
params_sin, covariancias_sin = curve_fit(sen, x ,y_sin)

print("Parâmetros da função seno:", params_sin)
```

```python
y_sin_fit = sen(x, *params_sin)

plt.figure(figsize=(12, 5))
plt.plot(x, y_sin, 'o', label='Dados')
plt.plot(x, y_sin_fit, '-', label='Ajuste Senoidal')
plt.legend()
plt.title('Senoidal')
```

```python
# Calculando o r^2 do modelo sen
ss_res_sin = np.sum((y_sin - y_sin_fit)**2)
ss_tot_sin = np.sum((y_sin - np.mean(y_sin))**2)
r2_sin = 1 - (ss_res_sin / ss_tot_sin)

print(f"R-squared sen: {r2_sin}")
```
